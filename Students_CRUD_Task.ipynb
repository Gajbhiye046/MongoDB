#### 3) Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

    { $cond: { if: <boolean-expression>, then: <true-case>, else: <false-case> } }
    
 or
 
    { $cond: [ <boolean-expression>, <true-case>, <false-case> ] }

for equality format

    { <field>: { $eq: <value> } }

```python
# url Reference : https://stackoverflow.com/questions/22107770/use-of-cond-and-eq-with-a-array-of-objects
# $switch url reference : https://www.mongodb.com/docs/manual/reference/operator/aggregation/switch/
lst =[]
pipeline_8 = [{ '$unwind': '$scores' },
              { "$project" :{  "_id" : 1,
                                "name":1,
                                "scores" :1,
                                 'Exam': {'$cond': [{'$eq': ['$scores.type', 'exam']},'$scores', 'null'] },
                                 'Quiz': {'$cond': [{'$eq': ['$scores.type', 'quiz']},'$scores','null']},
                                 'Homework': {'$cond':[{'$eq': ['$scores.type', 'homework']},'$scores','null']}
                   
                                     }},
              {'$group': {
                          '_id': '$_id',
                          'name':{'$first' : '$name'},
                          'Exam': { '$max': '$Exam' },
                          'Quiz': { '$max': '$Quiz' } ,
                          'Homework': { '$max': '$Homework' } 
                          
                     }},
              {'$project': {
                          '_id': 1,
                          'name':1,
                          'Exam': '$Exam',
                          'Quiz': '$Quiz',
                          'Homework': '$Homework'
                          
                     }},
              {
                '$project':
                  { '_id' : 1,
                    "name" : 1,
                   'Exam': '$Exam.score',
                   'Quiz': '$Quiz.score',
                   'Homework': '$Homework.score',
                    "Result" :
                    {
                      '$switch':
                        {
                          'branches': [
                            {
                              'case': { '$or' : [ { '$lte' : [ "$Exam.score" , 40 ] },
                                               { '$lte' : [ "$Quiz.score" , 40 ] },
                                               { '$lte' : [ "$Homework.score" , 40 ] } ] },
                             'then': "Fail."
                            },
                            {
                            'case': { '$and' : [ { '$gte' : [ "$Exam.score" , 75 ] },
                                               { '$gte' : [ "$Quiz.score" , 75 ] },
                                               { '$gte' : [ "$Homework.score" , 75 ] } ] },
                              'then': "Passed with distinction"
                            },
                          ],
                          'default': "Pass"
                        }
                     }
                  }
                },
              {'$sort' : {'name':1}}]
for i in Coll.aggregate(pipeline_8):
    lst.append(i)
```
